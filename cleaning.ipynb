```python
def clean_playoffs(file):
	table = pd.read_csv(file)
	teams_to_del = table.isna().sum()[3]
	temp_name_list = table["team_name"][teams_to_del:].reset_index()
	temp_name_list.dropna(inplace=True)
	table = table[:-teams_to_del].drop("Unnamed: 0", axis=1).reset_index()
	table["team_name"] = temp_name_list["team_name"]
	table["team_n"] = len(table)
	table["rank"] = np.arange(1, len(table) + 1)
	table.drop("index", axis=1, inplace=True)
	table.to_csv(file.replace(".csv", "_clean.csv"))
	display(table)
	return table
```
